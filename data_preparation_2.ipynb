**Qiymatlarni moslash(almashtirish) map()**

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/automobile_data_nan.csv")
df.head()
```

```python
df['num-of-cylinders'].unique()
```

```python
dict = {'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'eight':8, 'twelve':12}
df['cylinders'] = df['num-of-cylinders'].map(dict)
df
```

```python
df['price']
```

```python
kurs = 10720
df['price-uz'] = df['price'].map(lambda x:x*kurs)
df
```
