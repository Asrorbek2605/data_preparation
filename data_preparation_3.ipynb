```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/world_population.csv",index_col = "country")
df.head()
```

```python
df.loc['Uzbekistan']
```

```python
bins = [0, 1000, 10000, 30000, 50000, 100000, 300000, 1500000]
```

```python
population = df.pop2021
```

```python
band = pd.cut(population,bins)
band
```

```python
band.value_counts()
```

```python
df['band'] = band
df.head()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/automobile_data.csv",index_col = 0)
df.head()
```

```python
df.price.describe() #bizga malumot chiqarib beradi
```

```python
bins  = [0,10000,20000,50000]
```

```python
range = pd.cut(df.price,bins)
range
```

```python
range.value_counts()
```

```python
names =   ["arzon","orta","qimmat"]
range = pd.cut(df.price,bins,labels=names)
range.value_counts()
```

```python
range = pd.cut(df.price,3)
range.value_counts()
```

```python
range = pd.qcut(df.horsepower,3)
range.value_counts()
```

**Groupby yordamida guruhlash**

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/automobile_data.csv",index_col = 0)
df.head()
```

```python
df.company.value_counts()
```

```python
df.groupby('company').mean()
```

```python
df.groupby('company').max()
```

```python
df.groupby(['company','body-style']).price.max()
```

```python
df.groupby(['company']).agg([max,min,np.mean])
```
