**.fillna() - NaN qiymatlarni toldirish**

```python
df = pd.read_csv("https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/automobile_data_nan.csv")
df.head()
```

```python
df.fillna(0) # fillna metodi bizda NaN qiymatlarni 0ga aylantirib beradi
```

```python
df.fillna({'horsepower':np.mean(df['horsepower']),'wheel-base':1000000}) #bunda birdaniga bir nechta ustunlar bilan ishlash mumkin
```

```python
np.mean(df['horsepower']) #bunda horsepower ustunidagi qiymatlarni ortachasi hisoblab beradi.
```

```python
df
```

 **ffill** *Boshqasa usuli toldirishni*

```python
df.fillna(method='ffill') #ffill  metodi NaN qiymatli ni ozidan oldingi qiymati bilan joylashtiradi
```

**bfill** *Boshqasa usuli toldirishni*

```python
df.fillna(method = 'bfill')  #bfill  metodi NaN qiymatli ni ozidan keyingi qiymati  bilan joylashtiradi
```

```python
df[df.company == 'toyota'] #bunda faqat toyota ustunllarni chiqarib beradi
```

```python
df[df.company == 'toyota'].fillna(method = 'bfill')  #bunda faqat toyota ustunllarni chiqarib beradi va bfill methodi bajariladi
```

```python
df.fillna(df.mean()) #bunda barcha fillna qiymatlarni ustunlardagi ortacha qiymatlar bilan toldiradi
```
